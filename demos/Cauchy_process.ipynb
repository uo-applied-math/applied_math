# The Cauchy Process

Goal:
Let $N$ be the PPP on $[0,\infty) \times \mathbb{R}$
with mean intensity
$$
 \mu(dt dx) = \frac{dt dx}{|x|^2} . 
$$
Given $N$, let
$$
  C_t = \int_0^t \int_{-\infty}^\infty x N(dt dx) .
$$

First, how do we choose the $x$ coordinate?
$$
  \int_0^\infty \frac{1}{|x|^2} dx = \infty ,
$$
and so the number of points in any vertical strip in infinite.

Well,
$$
  \int_\epsilon^\infty \frac{1}{|x|^2} dx = \frac{1}{\epsilon} ,
$$
so the number of jumps of size bigger than $\epsilon$ is finite for any $\epsilon$.

We'll want to draw random variables $X$ that take values in $[\epsilon, \infty)$ and have probability
denssssity equal to $\epsilon dx / x^2$. (Then we can multiply these by a random $\pm 1$.)
Well,
$$\begin{aligned}
   f(x) &= \mathbb{P}\{ X > x \} \\
   &= \int_x^\infty \frac{\epsilon}{u^2} du \\
   &= \frac{\epsilon}{x} .
\end{aligned}$$
So, we can do this by letting $U$ be Unform on $[0,1]$ and $X = \epsilon/U$.

```python
import numpy as np
rng = np.random.default_rng()

epsilon = 0.001
t = 10
N = rng.poisson(lam=2 * t / epsilon)
X = epsilon / rng.uniform(size=N) * rng.choice([-1, 1], size=N)
T = rng.uniform(size=N) * t
T.sort()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,4))

ax.scatter(T, X);
```

```python
C = np.cumsum(X)

fig, ax = plt.subplots(figsize=(12,4))

ax.plot(T, C)
ax.set_xlabel("time")
ax.set_ylabel("C(t)");
```
