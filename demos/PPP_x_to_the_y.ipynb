# Poisson on a square

Simulate from the PPP on $[0,1]^2$ with mean intensity
$$
\mu(dx dxy) = 100 x^y dx dy .
$$

Plan:
1. Pick $N \sim \text{Poisson}(100 \log(2))$.
2. For $0 \le i < N$ let $U_i$ and $V_i$ be iid Uniform on $[0,1]$, and
3. let $$ Y_i = 2^{U_i} - 1, \qquad \text{and} $$
4. let $$ X_i = V_i^{1/(Y_i+1)} . $$
5. Enjoy.

```python
import numpy as np
rng = np.random.default_rng()
```

```python
N = rng.poisson(lam=10000 * np.log(2))
print(f"We will have {N} points!")
```

```python
U = rng.uniform(size=N)
V = rng.uniform(size=N)
Y = 2 ** U - 1
X = V ** (1/(Y + 1))
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,12))

ax.scatter(X, Y)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_aspect(1)
```
